```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load your weather-related data into a feature matrix (X)
# Ensure that X has rows as time periods (e.g., months) and columns as variables (rain, sun radiation, etc.)

# Replace the file path with your data file
weather_data = pd.read_excel('/Users/germankosenkov/Code projects/Indonesian S&D/May 10/Raw data/Weather data unscaled/After resampling/Resampled_Merged.xlsx')

# Separate the features (rain and sun radiation) from the target variable (palm oil production)
# Assuming 'Production' is the target variable column name
X = weather_data[[
    'Tn(I)', 'Tx(I)', 'Tavg(I)', 'RH_avg(I)', 'RR(I)', 'ss(I)', 'ff_x(I)', 'ddd_x(I)', 'ff_avg(I)',
    'Tn(J)', 'Tx(J)', 'Tavg(J)', 'RH_avg(J)', 'RR(J)', 'ss(J)', 'ff_x(J)', 'ddd_x(J)', 'ff_avg(J)',
    'Tn(N)', 'Tx(N)', 'Tavg(N)', 'RH_avg(N)', 'RR(N)', 'ss(N)', 'ff_x(N)', 'ddd_x(N)', 'ff_avg(N)',
    'Tn(O)', 'Tx(O)', 'Tavg(O)', 'RH_avg(O)', 'RR(O)', 'ss(O)', 'ff_x(O)', 'ddd_x(O)', 'ff_avg(O)',
    'Tn(S)', 'Tx(S)', 'Tavg(S)', 'RH_avg(S)', 'RR(S)', 'ss(S)', 'ff_x(S)', 'ddd_x(S)', 'ff_avg(S)'
]]
y = weather_data['Production']

#Load standard scaler and pca and then call transform with both 
#One file where you load scaler and then use transform 
#Store your scaler for other data 
#store in a folder whe

# Step 1: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Calculate the covariance matrix
covariance_matrix = np.cov(X_scaled.T)

# Step 3: Eigenvalue decomposition
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Step 4: Sort eigenvalues in descending order
eigenvalue_index_sorted = np.argsort(eigenvalues)[::-1]
eigenvalues_sorted = eigenvalues[eigenvalue_index_sorted]
eigenvectors_sorted = eigenvectors[:, eigenvalue_index_sorted]

# Step 5: Select the number of principal components
# You can choose a specific number or a percentage of explained variance
# For example, if you want to explain 95% of the variance:
total_variance = np.sum(eigenvalues_sorted)
explained_variance_ratio = eigenvalues_sorted / total_variance
cumulative_explained_variance_ratio = np.cumsum(explained_variance_ratio)

# Determine the number of components to retain
n_components_to_keep = np.argmax(cumulative_explained_variance_ratio >= 0.95) + 1

# Step 6: Apply PCA with the selected number of components
pca = PCA(n_components=n_components_to_keep)
X_pca = pca.fit_transform(X_scaled)

store.X_pca

# X_pca now contains the reduced-dimensional data with the top principal components

# Define the file path for the output Excel file
output_file_path = '/Users/germankosenkov/Code projects/Indonesian S&D/Production/Clean/Data/Refined_weather_data/4.Resampling/PCA.xlsx'  # You can change the file name as needed

# Create a DataFrame from X_pca
pca_data = pd.DataFrame(data=X_pca, columns=[f'PC{i}' for i in range(1, n_components_to_keep + 1)])

# Write the PCA data to an Excel file
pca_data.to_excel(output_file_path, index=False)

print(f'PCA data saved to {output_file_path}')
```
