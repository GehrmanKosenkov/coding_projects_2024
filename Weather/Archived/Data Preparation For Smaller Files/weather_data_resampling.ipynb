```python
import pandas as pd

data = '/Users/germankosenkov/Code projects/S&D estimation:forecasting/Indonesian S&D/June 17/Raw data/Weather data unscaled/Merged/Merged.xlsx'

df = pd.read_excel(data)

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')


df.set_index('Date', inplace=True)


monthly_df = df.resample('M').agg({
        'Tn(I)': 'mean',
        'Tn(J)': 'mean',
        'Tn(N)': 'mean',
        'Tn(O)': 'mean',
        'Tn(S)': 'mean',
        'Tx(I)': 'mean',
        'Tx(J)': 'mean',
        'Tx(N)': 'mean',
        'Tx(O)': 'mean',
        'Tx(S)': 'mean',
        'Tavg(I)': 'mean',
        'Tavg(J)': 'mean',
        'Tavg(N)': 'mean',
        'Tavg(O)': 'mean',
        'Tavg(S)': 'mean',
        'RH_avg(I)': 'mean',
        'RH_avg(J)': 'mean',
        'RH_avg(N)': 'mean',
        'RH_avg(O)': 'mean',
        'RH_avg(S)': 'mean',
        'RR(I)': 'sum',
        'RR(J)': 'sum',
        'RR(N)': 'sum',
        'RR(O)': 'sum',
        'RR(S)': 'sum',
        'ss(I)': 'sum',
        'ss(J)': 'sum',
        'ss(N)': 'sum',
        'ss(O)': 'sum',
        'ss(S)': 'sum',
        'ff_x(I)': 'mean',
        'ff_x(J)': 'mean',
        'ff_x(N)': 'mean',
        'ff_x(O)': 'mean',
        'ff_x(S)': 'mean',
        'ddd_x(I)': 'mean',
        'ddd_x(J)': 'mean',
        'ddd_x(N)': 'mean',
        'ddd_x(O)': 'mean',
        'ddd_x(S)': 'mean',
        'ff_avg(I)': 'mean',
        'ff_avg(J)': 'mean',
        'ff_avg(N)': 'mean',
        'ff_avg(O)': 'mean',
        'ff_avg(S)': 'mean',

    })



    # Reset the index to have a clean DataFrame
monthly_df.reset_index(inplace=True)

# Define the output file path for the current item
output_path = f'/Users/germankosenkov/Code projects/S&D estimation:forecasting/Indonesian S&D/June 17/Raw data/Weather data unscaled/After resampling/Resampled.xlsx'

# Write the modified DataFrame to the output file
monthly_df.to_excel(output_path, index=False)
```
